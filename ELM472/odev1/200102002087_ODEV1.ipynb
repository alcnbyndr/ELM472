```python
# ************ LIBRARIES ************
import numpy as np
import matplotlib.pyplot as plt

# ************ DATA SAMPLES ************
samples = np.load('samples.npy')
```

```python
X = samples[:,0]
Y = samples[:,1]

mean_x = np.mean(X)
mean_y = np.mean(Y)

m = len(X)

# I used necessary formula to calculate w1 and w2
numerator = 0
denominator = 0

for i in range(m):
    numerator   += (X[i] - mean_x) * (Y[i] - mean_y)
    denominator += (X[i] - mean_x) ** 2

# ************ W1 and W2 values calculated below ************    
m = numerator / denominator
c = mean_y - (m * mean_x)

print(f'm = {m} \nc = {c}')
```

```python
# ****** PLOTTING THE DATA AND REGRESSION LINE *******
max_x = np.max(X)
min_x = np.min(Y)

# Defining the values for the function
# It defines regression line's borders
x = np.linspace (min_x, max_x)
y = c + m * x

plt.plot(x, y, 'g', label = 'Regression Line')
plt.plot(X, Y, 'rx', label = 'Data given')

plt.xlabel('Data given in first column')
plt.ylabel('Data given in second column')
plt.legend()
plt.show()
```
